# How to use xarray-rrepr


```python
import xarray as xr

from xarray_rrepr import rrepr
```

```python
ds = xr.tutorial.load_dataset("air_temperature")
ds
```

```python
print(repr(ds))  # noqa: T201
```

```python
eval(repr(ds))  # noqa: S307
```

```python
print(rrepr(ds))  # noqa: T201
```

```python
eval(rrepr(ds))  # noqa: S307
```
